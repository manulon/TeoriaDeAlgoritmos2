```python
def mapper_func(k):
    return mapper[k]
```

```python
def contar_aristas(grafo):
  contador = 0
  for v in grafo:
    contador += len(list(grafo.neighbors(v)))
  return contador if nx.is_directed(grafo) else contador // 2

def proporcion_cruzan_campo(grafo, mapper=None):
  aristas_totales = contar_aristas(grafo)
  cruzan_bloque = 0
  visitados = set()
  mapper = mapper if mapper is not None else (lambda k: grafo[k]["type"])
  for v in grafo:
    for w in grafo.neighbors(v):
      if not nx.is_directed(grafo) and w in visitados:
        continue
      if mapper(v) != mapper(w):
        cruzan_bloque += 1
    visitados.add(v)
  return cruzan_bloque / aristas_totales

proporcion_real = round(proporcion_cruzan_campo(grafo, mapper_func), 2)
# Aca calcule la 'proporcion de cruce de campo', que es la homofilia global real, ya que es la
# proporcion de aristas que cruzan grupos sobre el total de aristas disponibles
```

```python
def proporcion_por_tipo(grafo, mapper=None):
  mapper = mapper if mapper is not None else (lambda k: grafo[k]["type"])
  cantidades = {}
  for v in grafo:
    cantidades[mapper(v)] = cantidades.get(mapper(v), 0) + 1
  props = {}
  for c in cantidades:
    props[c] = cantidades[c] / len(grafo)
  return props

proporciones = proporcion_por_tipo(grafo, mapper_func)
proporcion_teorica = 0
for clave, valor in proporciones.items():
  proporcion_teorica = proporcion_teorica + (proporciones[clave] * (1 - proporciones[clave]))

# Aca calcule la homofilia global teorica, que es la sumatoria de todas
# las probabilidades de agarrar un nodo y que pertenezca a un cierto grupo
```

```python
def proporcion_cruzan_campo_de_tipo(grafo, tipo, mapper=None):
  cruzan_bloque = 0
  visitados = set()
  aristas = 0
  mapper = mapper if mapper is not None else (lambda k: grafo[k]["type"])

  for v in grafo:
    if mapper(v) != tipo:
      continue
    for w in grafo.neighbors(v):
      aristas += 1
      if mapper(v) != mapper(w):
        cruzan_bloque += 1
    visitados.add(v)
  return cruzan_bloque / aristas

proporciones_cruce_campo = {}

for clave, valor in proporciones.items():
  if clave != None:
    proporciones_cruce_campo[clave] = round(proporcion_cruzan_campo_de_tipo(grafo, clave, mapper_func),2)
```

```python
print("Homofilia por grupos:")

valores_homofilia = []
claves_homofilia = []

for clave, valor in proporciones_cruce_campo.items():
  valores_homofilia.append(valor)
  claves_homofilia.append(clave)
  print(valor, '-', clave)
```

```python
def proporcion_cruzan_campo_de_tipo_con_pesos(grafo, tipo, mapper=None):
    cruzan_bloque = 0
    visitados = set()
    aristas = 0
    mapper = mapper if mapper is not None else (lambda k: grafo[k]["type"])

    for v in grafo:
        if mapper(v) != tipo:
            continue
        for w in grafo.neighbors(v):
            aristas += grafo.get_edge_data(v, w)['weight']
            if mapper(v) != mapper(w):
                cruzan_bloque += grafo.get_edge_data(v, w)['weight']
        visitados.add(v)
    return cruzan_bloque / aristas
```
